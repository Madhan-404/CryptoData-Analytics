**First order Differencing**

```python
series1 = BTC.Price.dropna() #original series
ax = plt.axes()
ax.set_facecolor('beige')
plt.plot(series1)
```

```python
ax = plt.axes()
series2 = BTC.Price.diff().dropna() #first order differenced series
ax.set_facecolor('silver')
plt.plot(series2)
```

Now, the time series is Stationary. So, we could Forecast now.

Differencing can help stabilise the mean of a time series by removing changes in the level of a time series, and therefore eliminating (or reducing) trend and seasonality.

```python
plt.rcParams.update({'figure.figsize':(12,7), 'figure.dpi':120})
# Original Series
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(series1); 
axes[0, 0].set_title('Original Series')
sgt.plot_acf(series1, ax=axes[0, 1])
# 1st Differencing
axes[1, 0].plot(series2); 
axes[1, 0].set_title('1st Order Differencing')
sgt.plot_acf(series2, ax=axes[1, 1])
plt.show()
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(series2); 
axes[0].set_title('1st Differencing')
axes[1].set(ylim=(-1,1))
sgt.plot_pacf(series2, ax=axes[1])
plt.show()
```

```python
#Find q in MA(q)
plt.rcParams.update({'figure.figsize':(12,3), 'figure.dpi':120})
fig, axes = plt.subplots(1, 2)
axes[0].plot(series2); 
axes[0].set_title('1st Differencing')
axes[1].set(ylim=(-1,1))
sgt.plot_acf(series2, ax=axes[1])
plt.show()
```

```python
model = ARIMA(BTC_train.Price, order=(1, 1, 1))
fitted = model.fit(disp=-1)  

# Forecast
fc, se, conf = fitted.forecast(len(BTC_test), alpha=0.05)  # 95% conf

# Make as pandas series
fc_series = pd.Series(fc, index=BTC_test.index)
lower_series = pd.Series(conf[:, 0], index=BTC_test.index)
upper_series = pd.Series(conf[:, 1], index=BTC_test.index)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(BTC_train.Price, label='training')
plt.plot(BTC_test.Price, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.15)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```
